```python
#Removing all the objects to only have numerical columns.
df1 =c.select_dtypes(exclude=['object'])
df1
```

```python
q1=df1.quantile(0.25)
q3=df1.quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
iqr=q3-q1
iqr
```

```python
#Finding the outliers & storing it in a variable
d=(df1<(q1-1.5*iqr))|(df1>(q3+1.5*iqr))
d.sum()
```

```python
#Removing the outliers from the original data frame
filter=c[~((df1<(q1-1.5*iqr))|(df1>(q3+1.5*iqr))).any(axis=1)]
filter
```
